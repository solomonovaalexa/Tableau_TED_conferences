## Исследование истории TED-конференций с помощью Tableau

TED (от англ. technology, education, design — «технологии, образование, дизайн») — некоммерческий фонд, который проводит популярные конференции. На них выступают специалисты из разных областей и читают лекции на актуальные социальные, культурные и научные темы.

Ссылка на презентацию: https://public.tableau.com/app/profile/alexandra.solomonova/viz/Book1_17225429653280/Story1?publish=yes

## Выводы на основе презентации

**Слайд "История выступлений"**

1) Чаще всего выступления проходили в США, Великобритании и Канаде;
2) Наиболее популярны следующие категории выступлений: технологии, общество, наука, глобальные проблемы и культура;
3) Наиболее популярная категория в 2020 и 2021 гг - глобальные проблемы.

**Слайд "Тематики выступлений"**

1) Наиболее популярны следующие категории выступлений: наука, технологии и бизнес;
2) Популярные категории почти не различаются в разных странах: В США, Канаде и Великобритании одинаково популярны выступления о науке, технологиях и культуре;
3) Аплодисменты у аудитории чаще вызывают спикеры со следующими тематиками выступлений: социальные изменения, образование и технологии;
4) Смех у аудитории чаще вызывают выступления на следующие тематики: образование, бизнес и дизайн;
5) Зависимости между длительностью выступления и количеством просмотров не наблюдается. Самое долгое выступление - "3 secrets to Netflix's success", длившееся более 60 мин. Самое просматриваемое выступление - "Do schools kill creativity?", набравшее более 70 млн просмотров.

**Слайд "Авторы выступлений"**

1) У авторов преобладают следующие области деятельности: писатели, журналисты и артисты;
2) На одного автора обычно приходится 1 выступление. Чаще всего выступал автор Juan Enriquez (8 раз);
3) Чаще остальных выступал дизайнер Tom Wujec (4 раза). Меньше всего смотрели его выступление под названием "Learn to use 13th century astrolabe" (менее 1 млн просмотров).

**Слайд "Языки выступлений"**

1) За анализируемый период подавляющее большинство выступлений велось на английском языке (более 300 тыс. выступлений);
2) В 2001 г все выступления TED были только на английском языке, а к 2021 г появились выступления на французском и испанском языках. На французском вели выступления преимущественно во Франции, а на испанском - в Аргентине.
